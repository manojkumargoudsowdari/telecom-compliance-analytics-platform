```python
import pandas as pd
```

```python
sample_url = "https://opendata.fcc.gov/resource/3xyp-aqkj.json?$limit=1000"
df = pd.read_json(sample_url)
df.head()
```

```python
df.shape
```

```python
df.columns.tolist()
```

```python
df.dtypes
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
selected_dimensions = ["company", "state", "issue", "timely_response"]

for column_name in selected_dimensions:
    print(f"\nTop values for {column_name}:")
    if column_name in df.columns:
        print(df[column_name].value_counts(dropna=False).head(10))
    else:
        print(f"Column '{column_name}' is not present in the FCC sample.")
```

```python
date_columns = ["ticket_created", "date_created", "issue_date"]

for column_name in date_columns:
    if column_name in df.columns:
        parsed_dates = pd.to_datetime(df[column_name], errors="coerce")
        print(f"\nDate coverage for {column_name}:")
        print({"min": parsed_dates.min(), "max": parsed_dates.max(), "non_null": parsed_dates.notna().sum()})
```
